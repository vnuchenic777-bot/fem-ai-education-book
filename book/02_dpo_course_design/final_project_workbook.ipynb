## 15. Экспорт результатов

Запустите следующую ячейку, чтобы сохранить таблицы в CSV и сформировать Markdown-файл с основными данными итогового проекта.


```python
def df_to_markdown(df: pd.DataFrame) -> str:
    """Простой экспорт DataFrame в Markdown без зависимости от tabulate."""
    headers = list(df.columns)
    rows = df.fillna("").astype(str).values.tolist()
    lines = []
    lines.append("| " + " | ".join(headers) + " |")
    lines.append("| " + " | ".join(["---"] * len(headers)) + " |")
    for row in rows:
        safe = [cell.replace("\n", "<br>").replace("|", "\\|") for cell in row]
        lines.append("| " + " | ".join(safe) + " |")
    return "\n".join(lines)

# CSV export
sections_df.to_csv(OUTPUT_DIR / "project_sections.csv", index=False)
rubric_df.to_csv(OUTPUT_DIR / "rubric.csv", index=False)
risks_df.to_csv(OUTPUT_DIR / "risk_register.csv", index=False)
technical_df.to_csv(OUTPUT_DIR / "technical_requirements.csv", index=False)
platform_flow_df.to_csv(OUTPUT_DIR / "platform_flow.csv", index=False)
checklist_df.to_csv(OUTPUT_DIR / "checklists.csv", index=False)

# Markdown export
report = []
report.append(f"# {PROGRAM_META['chapter']}\n")
report.append(f"**Курс:** {PROGRAM_META['course_title']}\n")
report.append("## Назначение итогового проекта\n")
for item in PROJECT_PURPOSES:
    report.append(f"- {item}")
report.append("\n## Структура итогового проекта\n")
report.append(df_to_markdown(sections_df))
report.append("\n## Рубрика оценивания\n")
report.append(df_to_markdown(rubric_df))
report.append("\n## Риски\n")
report.append(df_to_markdown(risks_df))
report.append("\n## Требования к цифровой фиксации\n")
report.append(df_to_markdown(technical_df))
report.append("\n## Цифровое сопровождение итогового проекта\n")
report.append(df_to_markdown(platform_flow_df))

markdown_path = OUTPUT_DIR / "08_FINAL_PROJECT_DPO_AI_generated.md"
markdown_path.write_text("\n".join(report), encoding="utf-8")

print(f"Готово. Файлы сохранены в папку: {OUTPUT_DIR}")
print(f"Markdown-отчет: {markdown_path}")
```
